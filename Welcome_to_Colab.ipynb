```
import pandas as pd

# Load data
df = pd.read_csv("heart_failure_prediction.csv")

# Count of null values BEFORE handling
print("üîç Null values before replacement:\n")
print(df.isnull().sum())

# Total missing values
total_missing_before = df.isnull().sum().sum()
print(f"\nüëâ Total missing values before replacement: {total_missing_before}\n")

# Replace nulls
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)

# Count of null values AFTER handling
print("‚úÖ Null values after replacement:\n")
print(df.isnull().sum())

# Total missing values
total_missing_after = df.isnull().sum().sum()
print(f"\n‚úÖ Total missing values after replacement: {total_missing_after}")
```

```
import pandas as pd

# Load data
df = pd.read_csv("heart_failure_prediction.csv")

# Count of null values BEFORE handling
print("üîç Null values before replacement:\n")
print(df.isnull().sum())

# Total missing values
total_missing_before = df.isnull().sum().sum()
print(f"\nüëâ Total missing values before replacement: {total_missing_before}\n")

# Replace nulls
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:
        df[column].fillna(df[column].mean(), inplace=True)
    else:
        df[column].fillna(df[column].mode()[0], inplace=True)

# Count of null values AFTER handling
print("‚úÖ Null values after replacement:\n")
print(df.isnull().sum())

# Total missing values
total_missing_after = df.isnull().sum().sum()
print(f"\n‚úÖ Total missing values after replacement: {total_missing_after}")
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# List of numeric features
numeric_cols = ["Age", "Resting_BP", "Cholesterol", "Max_Heart_Rate", "Oldpeak", "BMI"]

# Plot boxplots
plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, y=col, color="lightcoral")
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.suptitle("Boxplots for Outlier Detection", fontsize=16, y=1.02)
plt.show()
```

```
# Example: Identify outliers in 'Cholesterol'
Q1 = df['Cholesterol'].quantile(0.25)
Q3 = df['Cholesterol'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Cholesterol'] < Q1 - 1.5 * IQR) | (df['Cholesterol'] > Q3 + 1.5 * IQR)]
print(f"Number of outliers in 'Cholesterol': {outliers.shape[0]}")
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Check distribution of Heart Failure classes
print(df['Heart_Failure'].value_counts())
print("\nClass distribution (in %):")
print(df['Heart_Failure'].value_counts(normalize=True) * 100)

# Plot class distribution
sns.countplot(data=df, x='Heart_Failure', palette='Set2')
plt.title("Class Distribution: Heart Failure")
plt.xlabel("Heart Failure (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```
