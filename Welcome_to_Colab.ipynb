```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/heart_failure_prediction.csv")
```

```
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```
numeric_cols = ["Age", "Resting_BP", "Cholesterol", "Max_Heart_Rate", "Oldpeak", "BMI"]

df[numeric_cols].hist(bins=30, figsize=(15, 10), color='skyblue')
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Age", y="Max_Heart_Rate", hue="Heart_Failure", palette="coolwarm", alpha=0.6)
plt.title("Age vs Max Heart Rate by Heart Failure")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.grid(True)
plt.show()
```

```
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x="Gender", hue="Heart_Failure", palette="Set2")
plt.title("Heart Failure Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.subplot(1, 2, 2)
sns.countplot(data=df, x="Chest_Pain_Type", hue="Heart_Failure", palette="Set1")
plt.title("Heart Failure by Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(6, 6))
df['Heart_Failure'].value_counts().plot.pie(
    labels=["No", "Yes"], autopct='%1.1f%%', startangle=90, colors=["lightgreen", "salmon"]
)
plt.title("Heart Failure Distribution")
plt.ylabel("")
plt.show()
```

```
plt.figure(figsize=(12, 8))
corr = df[numeric_cols + ["Heart_Failure"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.show()
```
